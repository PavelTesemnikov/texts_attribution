# Функция для атрибуции текстов

```python
#Функция для построения дендрограмм по датафрейму с расстояниями
def get_dendrogram(df, title = ("", ""), save_graph = ""):
    
    ytdist = np.array(df)
    
    #Формируем массив из C_n^2 расстояний: [d(1,2), d(1,3), ..., d(1,n), d(2,3),..., d(2,n),...,d(n-1,n)]
    y1d = [ytdist[i,j] for i in range(len(ytdist)) for j in range(i+1, len(ytdist))]

    #Делаем подписи
    labels = list(df.index)
    
    #Это уже ответ
    Z = hierarchy.linkage(y1d, 'single')
    
    #А это рисунок
    plt.figure(figsize=(12,7))
    plt.title(title[0])
    dn = hierarchy.dendrogram(Z, labels = labels)
    
    if len(save_graph)>0:
        #print(1)
        plt.savefig(save_graph + "dendr_" + title[1] + '.png', bbox_inches='tight')
    
    return None

#Общая функция

def atribution(texts, dict_of_atribution_methods = None, dendrogram = True, heatmap = True, 
               table = True, save_graph = "", save_table = ""):
    
    if dict_of_atribution_methods is None:
        return None
    
    #Здесь общий словарь с итоговыми датафреймами
    dict_tables_result = {}
    
    #Перебираем все методы, чтобы заполнить словарь
    for key in dict_of_atribution_methods.keys():
        
        method = dict_of_atribution_methods[key]
        
        result = []
        
        #Запустили двойной цикл по всем текстам
        for key1 in tqdm(texts.keys()):
            res = []
            
            for key2 in texts.keys():
                
                #print(key1, key2, end = " ")
                
                text1 = texts[key1]
                text2 = texts[key2]
                res.append(method(text1, text2))
            result.append(res)
        
        #df -- итоговый датафрейм для этого метода
        df = pd.DataFrame(result, columns = texts.keys())
        df.index = texts.keys()
        
        #Дозаполнили словаоь с итоговыми датафреймами
        dict_tables_result.update({key: df})
    
    if dendrogram and heatmap:
        for key in dict_tables_result.keys():
            get_dendrogram(dict_tables_result[key], key, save_graph=save_graph)
            
            plt.figure(figsize=(14,5))
            plt.title(key[0])
            sb.heatmap(dict_tables_result[key])
            
            if len(save_graph) > 0:
                plt.savefig(save_graph + "heatmap_" + key[1] + '.png', bbox_inches='tight')
    
    elif heatmap:
        for key in dict_tables_result.keys():
            plt.figure(figsize=(14,5))
            plt.title(key[0])
            sb.heatmap(dict_tables_result[key])
            
            if len(save_graph) > 0:
                plt.savefig(save_graph + "heatmap_" + key[1] + '.png', bbox_inches='tight')
    
    elif dendrogram:
        for key in dict_tables_result.keys():
            get_dendrogram(dict_tables_result[key], key, save_graph=save_graph)
    
    if len(save_table) > 0:
        for key in dict_tables_result.keys():
            dict_tables_result[key].to_excel(excel_writer = save_table + key[1] + ".xlsx")
    
    if table:
        return dict_tables_result
```
